## 2
What is the cumulative distribution function (CDF) that corresponds to the following probability distribution function? Please state the value of the CDF for all possible values of $x$.

$f(x) = \begin{cases}
    \frac{1}{3} & 0 < x < 3 \\
    0           & \text{otherwise}
    \end{cases}$

**ANSWER**

CDF is defined as $F_X(x) = P(X \leq x)$, and the PDF is given that it is constant between 0 and 3 and is 0 otherwise. Therefore we can see that $F_X(x) = \int_{-\infty}^{x}p(t)dt = \int_{-\infty}^{x}\frac{1}{3}dt = \frac{1}{3}x$ between $\infty$ and 3.

Therefore, the CDF is:
$F(x) = \begin{cases}
    0           & x < 0 \\
    \frac{1}{3}x & 0 < x < 3 \\
    1           & x \geq 3
    \end{cases}$
